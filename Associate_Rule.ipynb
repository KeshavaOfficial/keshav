```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
!pip install mlxtend
```

```python
df= pd.read_csv('/content/anime.csv')
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support=0.07, use_colnames=True)
frequent_itemsets
```

```python

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
rules.sort_values('lift',ascending = False).head(10)
```

# 1. What is Lift and Why is it Important?
Lift measures the strength of an association between two items compared to random chance.
The formula for Lift is: Lift = Support(A → B) / (Support(A) × Support(B)).
A lift value greater than 1 indicates a strong relationship, while less than 1 suggests a weak or negative association.
It helps in market basket analysis, fraud detection, and recommendation systems.
A higher lift means items occur together more often than expected

# 2. What are Support and Confidence? How Do You Calculate Them?
Support shows how frequently an itemset appears in a dataset.
The formula for support is: Support(A) = (Transactions containing A) / (Total transactions).
Confidence measures the likelihood of one item being bought if another item is already bought.
The formula for confidence is: Confidence(A → B) = Support(A ∩ B) / Support(A).
Higher confidence means a stronger relationship between two items.

# 3. What are Some Limitations or Challenges of Association Rule Mining?
It can be computationally expensive, especially with large datasets.
Rare but important patterns may be ignored due to low support values.
Too many rules can make it difficult to find meaningful insights.
Some rules may be misleading or based on chance rather than real patterns.
Imbalanced datasets can affect the quality and accuracy of the generated rules.
