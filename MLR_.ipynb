```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = df.dropna(how='all')
```

# X as features affecting Price or X as Dependent Variables

```python
x = df.iloc[:,1:]
```

```python
x
```

# Y as independent Variable

```python
y = df.iloc[:,0]
```

```python
y
```

```python
df = pd.get_dummies(df, drop_first=True)
```

# Coverting categorical data to Numerical data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x.iloc[:,2] = le.fit_transform(x.iloc[:,2])
```

# imputing data in missing boxing using Mean Strategy

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x)
x = imputer.transform(x)
```

```python
x_imp = pd.DataFrame(x)
```

```python
x_imp
```

# train test split data


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Building Model

```python
from sklearn.linear_model import LinearRegression
MLR  = LinearRegression()
MLR.fit(x_train, y_train)

y_pred = MLR.predict(x_test)
```

```python
sns.scatterplot(x=y_test, y=MLR.predict(x_test))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Make predictions on the test set
y_pred = MLR.predict(x_test)

# Compute regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
ridge = Ridge(alpha=1.0)

ridge.fit(x_train, y_train)

y_pred_ridge = ridge.predict(x_test)

ridge_mse = mean_squared_error(y_test, y_pred_ridge)
ridge_r2 = r2_score(y_test, y_pred_ridge)
```

```python
print("Ridge Regression Performance:")
print(f"Mean Squared Error (MSE): {ridge_mse:.4f}")
print(f"R Score: {ridge_r2:.4f}")
```
